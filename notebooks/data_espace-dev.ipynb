# Les jeux de donnes Espace-Dev

DataSud : https://dataverse.ird.fr/dataverse/umr_espace-dev

```python
import requests
import pandas as pd
```

```python
url_datasud = "https://dataverse.ird.fr/"

nom_collection = "UMR ESPACE-DEV"

```

```python
try:
    reponse = requests.get(url_datasud + f"api/search?q=name_of_dataverse:{nom_collection}&type=dataset&per_page=300&show_facets=false&sort=date&order=desc").json()["data"]["items"]
except:
    print(requests.get(url_datasud + "/api/search?q=*&type=file&per_page=1&show_facets=true").json())
    reponse = {}

df_datasets = pd.json_normalize(reponse)
```

```python
df_datasets = df_datasets[df_datasets["name_of_dataverse"] == nom_collection]
```

À partir du champs contenant les informations de lieu `geographicCoverage` (sous forme de Json), nous faisons l'extraction des Pays et Ville.

Voici un exemple avec le jeu de données [10.23708/ACJTAW](https://dataverse.ird.fr/dataset.xhtml?persistentId=doi:10.23708/ACJTAW)

![Champs_geo](./../readme.files/datasud_geo.png)

```python
def extract_country_city(fields):
    # Check if fields is None or NaN, and return None for both columns if true
    if not isinstance(fields, list):
        return pd.Series([None, None])

    country = None
    city = None

    for item in fields:
        if 'country' in item:
            country = item['country']
        if 'city' in item:
            city = item['city']
    return pd.Series([country, city])

df_datasets[['country', 'city']] = df_datasets['geographicCoverage'].apply(extract_country_city)
```
